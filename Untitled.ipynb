```python
import pandas as pd
import numpy as np
import re
import gradio as gr
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df=pd.read_csv('collegereview2021.csv.zip')
```

```python
df.drop(df.columns[0:2],axis=1,inplace=True)
```

```python
df.head(20)
```

```python
import re
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)  # Remove special characters
    words = text.split()
    words = [w for w in words if w not in stop_words]  # Remove stopwords
    return ' '.join(words)


```

```python
df['review']=df['review'].apply(clean_text)
```

```python
# train test split

X=df['review']
Y=df['rating']
```

```python
# letter to number
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
tfidf = TfidfVectorizer()  # stop_words=None instead of 'english'


X_train_vec = tfidf.fit_transform(X_train)
X_test_vec = tfidf.transform(X_test)
```

```python
# train model

model=LinearRegression()
model.fit(X_train_vec,Y_train)
```

```python
def predict_rating(review_text):
    review = clean_text(review_text)  # Clean the text
    vec = tfidf.transform([review])  # TF-IDF transform
    prediction = model.predict(vec)[0]  # Predict rating
    return f"Predicted Rating: {round(prediction, 2)}"

```

```python
interface = gr.Interface(fn=predict_rating,
                         inputs="text",
                         outputs="text",
                         title="College Review Rating Predictor")

interface.launch()
```
